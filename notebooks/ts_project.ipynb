```python
def analyze_sentiment_roberta(text):
    """
    Analyze sentiment using RoBERTa - much more accurate for nuanced text
    """
    try:
        # Handle long lyrics by splitting into chunks
        if len(text) > 1000:
            chunks = [text[i:i+500] for i in range(0, len(text), 500)]
            sentiments = []
            for chunk in chunks[:3]:  # Analyze first 3 chunks max
                if len(chunk.strip()) > 10:
                    result = sentiment_pipeline(chunk)[0]
                    sentiments.append(result)

            # Aggregate results
            if sentiments:
                label_counts = {'positive': 0, 'negative': 0, 'neutral': 0}
                total_score = 0

                for sent in sentiments:
                    label_counts[sent['label']] += 1
                    total_score += sent['score'] if sent['label'] == 'positive' else -sent['score'] if sent['label'] == 'negative' else 0

                # Determine dominant sentiment
                dominant_label = max(label_counts, key=label_counts.get)
                avg_score = total_score / len(sentiments)

                return {
                    'roberta_label': dominant_label,
                    'roberta_confidence': round(avg_score, 4),
                    'roberta_compound': round(avg_score, 4),
                    'label_distribution': label_counts
                }

        # Standard analysis for shorter texts
        result = sentiment_pipeline(text[:1000])[0]  # Limit to first 1000 chars

        # Convert to numerical scores
        if result['label'] == 'positive':
            compound = result['score']
        elif result['label'] == 'negative':
            compound = -result['score']
        else:  # neutral
            compound = 0

        return {
            'roberta_label': result['label'],
            'roberta_confidence': round(result['score'], 4),
            'roberta_compound': round(compound, 4),
            'label_distribution': {result['label']: 1}
        }

    except Exception as e:
        print(f"❌ RoBERTa analysis error: {e}")
        return {
            'roberta_label': 'neutral',
            'roberta_confidence': 0,
            'roberta_compound': 0,
            'label_distribution': {'neutral': 1}
        }

def analyze_sentiment_textblob(text):
    """Keep TextBlob for comparison"""
    try:
        blob = TextBlob(text)
        return {
            'textblob_polarity': round(blob.sentiment.polarity, 4),
            'textblob_subjectivity': round(blob.sentiment.subjectivity, 4)
        }
    except:
        return {'textblob_polarity': 0, 'textblob_subjectivity': 0}

def extract_keyword_counts(text, keywords):
    """Count keyword occurrences"""
    counts = {}
    text_lower = text.lower()
    for keyword in keywords:
        pattern = r'\b' + re.escape(keyword) + r'\b'
        counts[keyword] = len(re.findall(pattern, text_lower))
    return counts

# Enhanced Taylor Swift keywords (same as before)
KEYWORDS = [
    # Time references
    'midnight', 'day', 'night', 'sun', 'moon', 'dark', 'light', 'dawn', 'dusk',
    'time', 'forever', 'never', 'always', 'memory', 'remember', 'forget',
    'hour', 'moment', 'yesterday', 'tomorrow', 'tonight', 'clock', 'timeless',

    # Emotions
    'love', 'heart', 'break', 'tears', 'smile', 'happy', 'sad', 'cry', 'pain',
    'angry', 'mad', 'hurt', 'scared', 'fear', 'hope', 'dream',
    'lonely', 'regret', 'sorry', 'shame', 'trust', 'betrayal', 'grace',

    # Colors
    'red', 'blue', 'golden', 'white', 'black', 'color', 'paint', 'gray',
    'silver', 'green', 'purple', 'pink', 'gold', 'blonde',

    # Weather & Nature
    'rain', 'storm', 'snow', 'winter', 'summer', 'spring', 'autumn',
    'wind', 'fire', 'water', 'ocean', 'river', 'sky', 'star',
    'earth', 'sea', 'lightning', 'sunshine', 'cloud', 'leaf', 'tree',

    # Relationships
    'boy', 'girl', 'man', 'woman', 'baby', 'darling', 'dear', 'sweet',
    'kiss', 'dance', 'touch', 'hold', 'leave', 'stay', 'wait',
    'lover', 'friend', 'enemy', 'ex', 'together', 'apart', 'promise'
]

print("✅ RoBERTa sentiment functions defined successfully!")
print(f"🔍 Tracking {len(KEYWORDS)} Taylor Swift keywords")
print("🧠 RoBERTa is state-of-the-art for nuanced text analysis!")
```

```python
print("🎵 PROCESSING TAYLOR SWIFT SONGS WITH RoBERTa ANALYSIS...")
print("=" * 60)

processed_songs = []
analysis_stats = {'positive': 0, 'negative': 0, 'neutral': 0}

for i, song in enumerate(songs_data):
    if i % 10 == 0:  # Progress update every 10 songs
        print(f"🔍 Analyzed {i}/{len(songs_data)} songs...")

    # Get RoBERTa sentiment scores
    roberta_sentiment = analyze_sentiment_roberta(song['lyrics'])
    textblob_sentiment = analyze_sentiment_textblob(song['lyrics'])
    keyword_counts = extract_keyword_counts(song['lyrics'], KEYWORDS)

    # Update stats
    analysis_stats[roberta_sentiment['roberta_label']] += 1

    # Create processed song entry
    processed_song = {
        'title': song['title'],
        'album': song['album'],
        'year': song['year'],
        'lyrics': song['lyrics'],
        'word_count': len(song['lyrics'].split()),
        'character_count': len(song['lyrics']),
        # RoBERTa sentiment scores
        'roberta_label': roberta_sentiment['roberta_label'],
        'roberta_confidence': roberta_sentiment['roberta_confidence'],
        'roberta_compound': roberta_sentiment['roberta_compound'],
        'label_distribution': roberta_sentiment['label_distribution'],
        # TextBlob for comparison
        'textblob_polarity': textblob_sentiment['textblob_polarity'],
        'textblob_subjectivity': textblob_sentiment['textblob_subjectivity']
    }

    # Add keyword counts
    processed_song.update(keyword_counts)
    processed_songs.append(processed_song)

print("=" * 60)
print(f"✅ SUCCESS: Processed {len(processed_songs)} songs with RoBERTa!")
print(f"📊 RoBERTa Sentiment Distribution:")
for sentiment, count in analysis_stats.items():
    percentage = (count / len(processed_songs)) * 100
    print(f"   {sentiment.upper()}: {count} songs ({percentage:.1f}%)")
```

```python
# Create DataFrame from processed songs
df = pd.DataFrame(processed_songs)

# Convert RoBERTa labels to categorical for analysis
df['roberta_sentiment'] = df['roberta_label']

print("📊 RoBERTa SENTIMENT ANALYSIS RESULTS")
print("=" * 50)
print(f"🎵 Total songs analyzed: {len(df)}")
print(f"📀 Albums: {df['album'].nunique()}")
print(f"📅 Year range: {df['year'].min()} - {df['year'].max()}")

print(f"\n🎯 RoBERTa SENTIMENT BREAKDOWN:")
sentiment_counts = df['roberta_label'].value_counts()
for sentiment, count in sentiment_counts.items():
    percentage = (count / len(df)) * 100
    confidence_avg = df[df['roberta_label'] == sentiment]['roberta_confidence'].mean()
    print(f"  {sentiment.upper()}: {count} songs ({percentage:.1f}%)")
    print(f"    Average Confidence: {confidence_avg:.3f}")

print(f"\n📈 RoBERTa COMPOUND SCORES:")
print(f"  Range: {df['roberta_compound'].min():.3f} to {df['roberta_compound'].max():.3f}")
print(f"  Average: {df['roberta_compound'].mean():.3f}")
print(f"  Std Dev: {df['roberta_compound'].std():.3f}")

print(f"\n🔍 TEXTBLOB COMPARISON:")
print(f"  Average Polarity: {df['textblob_polarity'].mean():.3f}")
print(f"  Average Subjectivity: {df['textblob_subjectivity'].mean():.3f}")

print(f"\n🔤 TOP KEYWORDS FOUND:")
keyword_totals = {keyword: df[keyword].sum() for keyword in KEYWORDS}
top_keywords = sorted([(k, v) for k, v in keyword_totals.items() if v > 0],
                     key=lambda x: x[1], reverse=True)[:10]

for keyword, count in top_keywords:
    print(f"  {keyword}: {count} occurrences")
```

```python
print("🎭 RoBERTa DEEP DIVE ANALYSIS")
print("=" * 45)

# Most positive and negative songs
most_positive = df.loc[df['roberta_compound'].idxmax()]
most_negative = df.loc[df['roberta_compound'].idxmin()]

print(f"🎉 MOST POSITIVE SONG:")
print(f"   '{most_positive['title']}' from {most_positive['album']}")
print(f"   RoBERTa Label: {most_positive['roberta_label']}")
print(f"   Confidence: {most_positive['roberta_confidence']:.3f}")
print(f"   Compound Score: {most_positive['roberta_compound']:.3f}")

print(f"\n😢 MOST NEGATIVE SONG:")
print(f"   '{most_negative['title']}' from {most_negative['album']}")
print(f"   RoBERTa Label: {most_negative['roberta_label']}")
print(f"   Confidence: {most_negative['roberta_confidence']:.3f}")
print(f"   Compound Score: {most_negative['roberta_compound']:.3f}")

print(f"\n📀 ALBUM SENTIMENT ANALYSIS:")
album_analysis = df.groupby('album').agg({
    'roberta_compound': 'mean',
    'roberta_confidence': 'mean',
    'title': 'count'
}).round(3)

for album, row in album_analysis.iterrows():
    compound = row['roberta_compound']
    if compound > 0.3:
        emoji = "😊"
        mood = "Very Positive"
    elif compound > 0.1:
        emoji = "🙂"
        mood = "Positive"
    elif compound > -0.1:
        emoji = "😐"
        mood = "Neutral"
    elif compound > -0.3:
        emoji = "😟"
        mood = "Negative"
    else:
        emoji = "😢"
        mood = "Very Negative"

    print(f"  {emoji} {album}: {compound} ({mood}) - {int(row['title'])} songs")

print(f"\n📅 YEARLY SENTIMENT TRENDS:")
yearly_trends = df.groupby('year')['roberta_compound'].mean().round(3)
for year, sentiment in yearly_trends.items():
    if sentiment > 0.1:
        trend = "📈 Rising"
    elif sentiment < -0.1:
        trend = "📉 Falling"
    else:
        trend = "➡️ Stable"
    print(f"  {year}: {sentiment} {trend}")
```
